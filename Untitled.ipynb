```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
csv_file_path = 'cpi_inflation.csv'
data = pd.read_csv(csv_file_path)

# Extract columns
t = data['t']
realized = data['realized']

# Creating dummy data for IT, ITup, and ITlo for demonstration
# Replace these with actual columns if present in the dataset
IT = np.linspace(1, 5, len(t))  # Replace with `data['IT']` if available
ITup = IT + 0.5
ITlo = IT - 0.5

# Create xfill and yfill for shaded region
xfill = np.concatenate([t, t[::-1]])
yfill = np.concatenate([ITup, ITlo[::-1]])

# Define MATLAB-like color and line style mappings
color_styles = [
    ('b', '--', 'x'),  # blue dashed with 'x' marker
    ('c', '-.', '^'),  # cyan dash-dot with '^' marker
]
# Generate the list of styles for all dynamic columns
dynamic_styles = [color_styles[i % len(color_styles)] for i in range(len(data.columns) - 2)]


# Identify dynamic columns (all except `t` and `realized`)
dynamic_columns = [col for col in data.columns if col not in ['t', 'realized']]

# --- Plot with Dynamic Columns ---
plt.figure(figsize=(10, 5))
fill_color = [0.9, 0.9, 0.9]
plt.fill(xfill, yfill, color=fill_color, label='Target range')

plt.plot(t, IT, 'k:', label='Target point')
plt.plot(t, realized, 'r-o', label='Actual inflation rate')

# Add dynamic columns to the plot
for i, col in enumerate(dynamic_columns):
    color, linestyle, marker = dynamic_styles[i]
    # Ensure there are no missing values before plotting
    if data[col].notnull().any():
        plt.plot(t, data[col], linestyle=linestyle, color=color, marker=marker, label=col)



# Example: Add more series if necessary
# Additional variables can be added here, using columns like `data['VarNameX']`

plt.xlim([2012, 2026])
plt.ylim([0, 6])
plt.xticks(range(2012, 2027))
plt.yticks(range(0, 7))
plt.title('CPI Inflation Rate')
plt.show()

# --- Second Plot ---
plt.figure(figsize=(10, 5))
plt.fill(xfill, yfill, color=fill_color, label='Target range')

# Using dummy data for `core` and `cpi` for demonstration
# Replace these with actual columns from the dataset if available
core = np.sin(np.linspace(0, 2 * np.pi, len(t))) + 2.5  # Replace with `data['core']`
cpi = np.cos(np.linspace(0, 2 * np.pi, len(t))) + 2.5  # Replace with `data['cpi']`

plt.plot(t, core, 'b-.', label='Core Inflation')
plt.plot(t, cpi, 'r', label='CPI Inflation')
plt.plot(t, IT, 'k--', label='Target point')

plt.xlim([1999, 2025])
plt.ylim([0, 6])
plt.title('Inflation Targets in Korea')
plt.legend(loc='upper left')
plt.show()
```
